```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('houseSmallData.csv', index_col=0)
display(df)
print(df.dtypes)
print(df.columns)
print(df.shape)
display(df.describe())
df = df.select_dtypes(include=[np.number])
display(df.isnull().sum())

df = df.interpolate().dropna(axis=1)

sns.histplot(data=df, x="SalePrice")
plt.grid()
plt.savefig('sale_price_hist.png')
plt.show()

display(df['SalePrice'].mean())
display(df['SalePrice'].std())
```

```python
df_new = df
df_new['SalePriceLog'] = np.log(df['SalePrice'])
sns.histplot(data=df_new, x="SalePriceLog")
plt.grid()
plt.savefig('sale_price_hist_log.png')
plt.show()

display(df['SalePrice'].skew())
display(np.log(df['SalePrice']).skew())
```

```python
# Data Correlation
tbl = df.corr()

display(tbl.sort_values('SalePrice', ascending=False)['SalePrice'])
best_corr = tbl.sort_values('SalePrice', ascending=False)['SalePrice'].index[0:11]
display(best_corr)
```

```python
# Keep Relevant columns
df_best = df[best_corr]

# Show Data Types
display(df_best.dtypes)

# Show # of NaN
display(df_best.isna().sum())

```

```python
# Remove all prices with NaN
# df_best['SalePrice'] = df_best['SalePrice'].dropna()
df_best['SalePrice'] = df_best['SalePrice'].dropna()
df_best['GarageYrBlt'].fillna(df_best['GarageYrBlt'].mean(), inplace = True)

# # Casting as Category Data
# df["MSZoning"] = df["MSZoning"].astype("category")
# df["SaleType"] = df["SaleType"].astype("category")
# df["SaleCondition"] = df["SaleCondition"].astype("category")
# df['BldgType'] = df["BldgType"].astype("category")
# df['CentralAir'] = df["CentralAir"].astype("category")
# df['Neighborhood'] = df["Neighborhood"].astype("category")
# df['HouseStyle'] = df["HouseStyle"].astype("category")
```

```python
# Plot Variables
for val in best_corr:
    if val != 'SalePrice':
        sns.regplot(x = val,y = 'SalePrice', data=df_best)
        plt.legend()
        plt.grid()
        plt.ylabel('Price ($)')
        plt.savefig(val + '.png')
        plt.show()
        
# plt.scatter(df['LotArea'],df['SalePrice'])
# GarageCars, OverallCond, YearBuilt, LotArea, TotRmsAbvGrd, BldgType, HouseStyle, TotalBsmtSF, 1stFlrSF, 2ndFlrSF
```

```python
# Features/Target
target = df_best['SalePrice']
data = df_best.drop(['SalePrice'], axis=1)

# Linear Model
linear_regression = LinearRegression()
linear_regression.fit(data, target)

# Result
display(linear_regression.coef_)
display(linear_regression.intercept_)

# Prediction
target_pred = linear_regression.predict(data)

# Printing
sns.scatterplot(x=target_pred, y=target, color="black", alpha=0.5)
plt.plot(target, target)
plt.title("Prediction vs. Model")
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.grid()
plt.show()
```

```python
mae = mean_absolute_error(target, target_pred)
mse = mean_squared_error(target, target_pred)
rmse = np.sqrt(mse)
print(f'Mean absolute error: {mae:.2f}')
print(f'Mean squared error: {mse:.2f}')
print(f'Root mean squared error: {rmse:.2f}')
```

```python
actual_minus_predicted = sum((target - target_pred)**2)
actual_minus_actual_mean = sum((target - target.mean())**2)
r2 = 1 - actual_minus_predicted/actual_minus_actual_mean
print('R²:', r2)
print(f'R²: {linear_regression.score(data, target)}')
```
